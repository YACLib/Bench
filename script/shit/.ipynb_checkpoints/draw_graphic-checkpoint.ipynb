```python
import csv
file = open("results.csv")

csvreader = csv.reader(file)
header = next(csvreader)
print(header)
rows = {'stl': [], 'folly': [], 'yaclib': [], 'qt': []}
for row in csvreader:
    time = float(row[1].strip()[0:-2])
    if row[0].lower().startswith('stl'):
        rows['stl'].append(time)
    elif row[0].lower().startswith('folly'):
        rows['folly'].append(time)
    elif row[0].lower().startswith('yaclib'):
        rows['yaclib'].append(time)
    elif row[0].lower().startswith('qt'):
        rows['qt'].append(time)
file.close()

print(rows)
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5)

for name, results in rows.items():
    x = np.linspace(0, len(results), len(results))
    y = results
    print(name)
    line  = ax.plot(x, y, label=name)


ax.legend()
plt.show()
fig.savefig('results.png', dpi=100)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# make data:
x_max = max(map(len, rows.values()))

x = np.arange(0, x_max * 4, 4)
y = rows['yaclib']
y2 = rows['folly']
y3 = rows['qt']

for i in range(len(y)):
    max_el = max(y[i], y2[i], y3[i])
    y[i] /= max_el
    y2[i] /= max_el
    y3[i] /= max_el

width = 1

fig, ax = plt.subplots()
fig.set_size_inches(14.5, 10)

ax.bar(x, y, width, edgecolor="white", linewidth=1, label='yaclib')
ax.bar(x + width, y2, width, edgecolor="white", linewidth=1, label = 'folly')
ax.bar(x + 2* width, y3, width, edgecolor="white", linewidth=1, label='qt')

ax.set(xticks=np.arange(0, 0))
ax.set_title(' ', fontsize='36')
ax.set_ylabel("Ratio", fontsize='18')

ax.legend(prop={'size': 20}, loc='lower right')
plt.show()
fig.savefig('results.png', dpi=100)
```
